# Exploration of the dataset of funerary inscriptions

The most represented province in the dataset is Roma (68,368) followed by Africa proconsularis (16,984) and Numidia (11,314). A plot shows the distribution of the inscriptions over the provinces.

The most frequent raw_spantime is 301 to 500 with 12,975 followed by 101 to 200 with 9,676 inscriptions.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##open the dataset of funerary inscriptions (172,958 rows)
Inscriptions = pd.read_csv("/Users/u0154817/OneDrive - KU Leuven/Documents/ICLL Prague June 2023/Output/Tituli_Sepulcrales_new.csv")
```

```python
len(Inscriptions)
```

```python
Inscriptions["province"].value_counts()
```

```python
# Calculate the value counts
provinces = Inscriptions["province"].value_counts()

# Exclude the highest value
value_counts_excluded = provinces.iloc[1:]

## sort the counts dictionary by key in ascending order
value_counts_excluded = dict(sorted(value_counts_excluded.items()))

plt.figure(figsize=(15, 5))


## plot the sorted counts as a bar plot
plt.bar(value_counts_excluded.keys(), value_counts_excluded.values())
plt.xticks(rotation=90)

plt.xlabel('province')
plt.ylabel('n of Latin funerary inscriptions')

## show the plot
plt.show()
```
